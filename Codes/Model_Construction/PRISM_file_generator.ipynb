```python
# Creating states and identifying unique states in the model.

state_seq=[]
for i in range(len(t)):
    state_sequence=[]
    for j in range(len(t[i])):
        d=''
        for k in range(len(t[i][j])):
            d+=t[i][j][k]+'___'
        d = d[:-3]
        if(j==0):
            d+='___onest'
        state_sequence.append(d)
    state_seq.append(state_sequence)

state_un=[]
for i in range(len(state_seq)):
    for j in range(len(state_seq[i])):
        if state_seq[i][j] in state_un:
            continue
        else:
            state_un.append(state_seq[i][j])
```

```python
# Calculating the degree of each state in the model and storing it in a dictionary.

degree={}
for i in range(len(state_seq)):
    for j in range(len(state_seq[i])):
        if 'CLOSING' in state_seq[i][j].split('___'):
            if state_seq[i][j] in degree:
                continue
            else:
                degree[state_seq[i][j]]=1
                continue
        if state_seq[i][j] in degree:
            degree[state_seq[i][j]]+=1
        else:
            degree[state_seq[i][j]]=1
```

```python
# Identifying the initial and final state for each case and carefully storing
# it to relate it to the initial and final state of the model.

PRISM_state=[]
ini=['0']
ini.append(len(t))
ini.append(['INITIAL'])
ini.append('INITIAL')
PRISM_state.append(ini)
for i in range(len(state_un)):
    j=i+1
    a=[]
    a.append(str(j))
    label=[]
    l=list(state_un[i])
    idx=[-1]
    for j in range(len(l)):
        if l[j]=='_' and l[j+1]=='_' and l[j+2]=='_':
            idx.append(j)
    lab=[]
    for j in range(len(idx)-1):
        laba=''
        for k in range(idx[j]+3,idx[j+1]):
            laba+=l[k]
        if laba[0]=='_':
            laba=''.join(('j_',laba))
        if laba=='EQ':
            laba=''.join(('NR',laba))
        lab.append(laba)
    laba=''
    for j in range(idx[-1]+3,len(l)):
        laba+=l[j]
    lab.append(laba)
    for j in range(len(lab)):
        label.append(lab[j])
    deg=degree[state_un[i]]
    a.append(deg)
    a.append(label)
    a.append(state_un[i])
    PRISM_state.append(a)
skip_id=[]
for i in range(len(PRISM_state)):
    if 'CLOSING' in PRISM_state[i][-2]:
        skip_id.append(i)
fin=[str(len(PRISM_state))]
fin.append(1)
fin.append(['FINAL'])
fin.append('FINAL')
PRISM_state.append(fin)
```

```python
# Creating the state sequence for the PRISM model and also creating the adjacency dictionary for each state.

PRISM_state_seq=[]
for i in range(len(state_seq)):
    for j in range(len(state_seq[i])):
        for k in range(len(PRISM_state)):
            if state_seq[i][j]==PRISM_state[k][-1]:
                PRISM_state_seq.append(PRISM_state[k][0])
                break

PRISM_adjacency={}
for i in range(len(PRISM_state_seq)-1):
    if int(PRISM_state_seq[i]) in skip_id:
        continue
    if (PRISM_state_seq[i],PRISM_state_seq[i+1]) in PRISM_adjacency:
        PRISM_adjacency[PRISM_state_seq[i],PRISM_state_seq[i+1]]+=1
    else:
        PRISM_adjacency[PRISM_state_seq[i],PRISM_state_seq[i+1]]=1
PRISM_adjacency_list=list(PRISM_adjacency)
PRISM_adjacency_list.sort()

PRISM_code=[]
for i in range(len(PRISM_state)):
    a=[]
    a.append(PRISM_state[i][0])
    a.append(PRISM_state[i][1])
    PRISM_code.append(a)

if PRISM_code[0][0]=='0':
    for i in range(len(PRISM_state)):
        b=[]
        if 'onest' in PRISM_state[i][-1]:
            for j in range(len(PRISM_code)):
                if PRISM_state[i][0] == PRISM_code[j][0]:
                    num=PRISM_code[j][1]
            b.append(PRISM_state[i][0])
            b.append(num)
            PRISM_code[0].append(b)
for i in range(len(PRISM_code)):
    if int(PRISM_code[i][0]) in skip_id:
        for j in range(len(PRISM_state)):
            c=[]
            if 'FINAL' in PRISM_state[j][-1]:
                c.append(PRISM_state[j][0])
                c.append(PRISM_code[i][1])
                PRISM_code[i].append(c)
for i in range(len(PRISM_code)):
    for j in range(len(PRISM_adjacency)):
        a=[]
        if PRISM_code[i][0]==PRISM_adjacency_list[j][0]:
            a.append(PRISM_adjacency_list[j][1])
            a.append(PRISM_adjacency[PRISM_code[i][0],PRISM_adjacency_list[j][1]])
            PRISM_code[i].append(a)
```

```python
# We divide each state-count by the degree of the state to obtain the probability and also create a set of 
# unique labels in the model to help us with the PRISM code.

for i in range(len(PRISM_code)):
    for j in range(2,len(PRISM_code[i])):
        PRISM_code[i][j][1]=round(float(PRISM_code[i][j][1])/float(PRISM_code[i][1]),8)

unique_labels=[]
for i in range(len(PRISM_state)):
    for x in PRISM_state[i][2]:
        if x not in unique_labels:
            unique_labels.append(x)
#print(len(unique_labels))
#print(unique_labels)
```
